# Principal Component Analysis
An Unsupervised Learning technique used to examine the interrelations among a set of variables in order to identify the underlying structure of those variables. Also Known as a general factor analysis
It determines several orthogonal(Right Angles) lines of best fit as opposed to regression that finds the best fit. Thes elines are perpendicular to each other in n-dimensional space. n-dimensional is the variable sample sapce. That means there are as many dimensions as there are variables, i.e if there are 4 variables in a dataset, the sample space is 4-dimensional.



```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
##load data
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
print(cancer['target_names'])
```

```python
#create a DataFrame from the dataset
df = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
```

```python
df.head()
```

```python
#PCA with scikit-learn
#Start with Standard Scaler
from sklearn.preprocessing import StandardScaler
```

```python
#scale data so each features has a single unit variance before we do PCA
#create an instance of the StandardScaler
scaler = StandardScaler()
```

```python
#fit data to scaler
scaler.fit(df) #df is just the features, not target here, that is why we use the whole thing. 
```

```python
#transform data
scaled_data = scaler.transform(df)
```

```python
#PCA goes through a similar process as the other preprocessing funtions, also specify how many components we want to keep we creating the pca instance
#create an instance on the pca class
#find principal components using the fit method, bascially fit the df to the pca object
#apply the rotaion and dimensionality reduction by calling transform

from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2) #specifying number of components we want to keep
```

```python
#fit transformed dat to the pca, in this case it's scaled_data
pca.fit(scaled_data)
```

```python
#transform the datab to its first principal components
x_pca = pca.transform(scaled_data)
```

```python
#check the shape pf the old data
scaled_data.shape #has all the components
```

```python
#shape after the data is transform with the pca object
x_pca.shape #since we set the number of pca components to 2 when we creating the pca object, our new transformed data has the first two components
```

```python
#plot the x_pca data to see the results
plt.figure(figsize=(10,6))
#grab all the rows from column zero for x, 1 for y, color based on the target column, color map of rainbow
plt.scatter(x_pca[:,0],x_pca[:,1], c=cancer['target'], cmap='rainbow') 
plt.xlabel('First Principal Compoenent')
plt.ylabel('Second Principal Compoenent')
```

```python
#plt scatter plots with both legends showing
plt.figure(figsize=(10,6))

# Plot each target class separately
for label, color in zip([0, 1], ['red', 'blue']):
    plt.scatter(
        x_pca[cancer['target'] == label, 0],
        x_pca[cancer['target'] == label, 1],
        label=cancer['target_names'][label],
        alpha=0.6,
        c=color
    )

plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.legend()
```

```python
#these components do not relate one to one to a specific feature on the data
#they correspond to a combination of the actual features
pca.components_
```

```python
#lets visualize the pca data with the principal components
#conver to a data frame
df_comp = pd.DataFrame(pca.components_, columns=cancer['feature_names'])
```

```python
df_comp
```

```python
#vsiualize with a heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df_comp, cmap='plasma')
```
